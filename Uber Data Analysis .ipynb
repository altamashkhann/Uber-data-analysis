```python
#Success Rate by Vehicle Type:

import pandas as pd

# Total bookings per vehicle
total_by_vehicle = df.groupby('Vehicle Type')['Booking ID'].nunique()

# Completed bookings per vehicle
completed_by_vehicle = df[
    df['Booking Status'] == 'Completed'
].groupby('Vehicle Type')['Booking ID'].nunique()

# Combine and calculate success rate
success_rate_by_vehicle = (
    completed_by_vehicle / total_by_vehicle
).fillna(0) * 100
```

```python
success_rate_df = success_rate_by_vehicle.reset_index()
success_rate_df.columns = ['Vehicle Type', 'Success Rate (%)']

ax = sns.barplot(data=success_rate_df, x='Vehicle Type', y='Success Rate (%)')

ax.bar_label(ax.containers[0], fmt='%.1f%%')

plt.title('Success Rate by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Success Rate (%)')
plt.ylim(0, 100)
plt.show()
```

# Dashboard: 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure datetime
df['Time'] = pd.to_datetime(df['Time'])

# Hour extraction
df['Hour'] = df['Time'].dt.hour

# Total bookings by vehicle
vehicle_counts = df.groupby('Vehicle Type')['Booking ID'].nunique().reset_index()
vehicle_counts.columns = ['Vehicle Type', 'Booking ID']

# Booking status distribution
status_counts = df.groupby('Booking Status')['Booking ID'].nunique().reset_index()
status_counts.columns = ['Booking Status', 'Count']

# Success rate by vehicle
total_by_vehicle = df.groupby('Vehicle Type')['Booking ID'].nunique()
completed_by_vehicle = df[df['Booking Status'] == 'Completed'] \
    .groupby('Vehicle Type')['Booking ID'].nunique()

success_rate = (completed_by_vehicle / total_by_vehicle * 100).fillna(0).reset_index()
success_rate.columns = ['Vehicle Type', 'Success Rate %']

# Bookings by hour
hourly_bookings = df.groupby('Hour')['Booking ID'].nunique().reset_index()
hourly_bookings.columns = ['Hour', 'Booking ID']

#I have already had this, so I used chatgpt to sum-up and create all in one to not mess up after.
```

```python
#Dashboard Layout:

[ Plot 1 | Plot 2 ]
[ Plot 3 | Plot 4 ]
```

```python
#Dashboard figure size:

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

#plot 1:

sns.barplot(data=vehicle_counts, x='Vehicle Type', y='Booking ID', ax=axes[0, 0])

axes[0, 0].set_title('Total Bookings by Vehicle Type')
axes[0, 0].set_xlabel('Vehicle Type')
axes[0, 0].set_ylabel('Total Bookings')

#plot 2:
sns.barplot(data=status_counts, x='Booking Status', y='Count', ax=axes[0, 1])

axes[0, 1].set_title('Booking Status Distribution')
axes[0, 1].set_xlabel('Booking Status')
axes[0, 1].set_ylabel('Number of Bookings')

#plot 3:
ax = sns.barplot(data=success_rate, x='Vehicle Type', y='Success Rate %', ax=axes[1, 0])

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)
    
axes[1, 0].set_title('Success Rate by Vehicle Type')
axes[1, 0].set_ylim(0, 100)
axes[1, 0].set_xlabel('Vehicle Type')
axes[1, 0].set_ylabel('Success Rate %')

#plot 4:
sns.lineplot(data=hourly_bookings, x='Hour', y='Booking ID', marker='o', ax=axes[1, 1])

axes[1, 1].set_title('Bookings by Hour of Day')
axes[1, 1].set_xlabel('Hour of Day')
axes[1, 1].set_ylabel('Total Bookings')
axes[1, 1].set_xticks(range(0, 24))

plt.tight_layout()
plt.show()
```

```python
#Interactive Dashboard using same metrics:

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
#Dashboard Figure layout:
fig = make_subplots(
    rows=2, cols=2,
    specs=[
        [{"type": "xy"}, {"type": "domain"}],   # row 1
        [{"type": "xy"}, {"type": "xy"}]        # row 2
    ],
    subplot_titles=(
        "Total Bookings by Vehicle Type",
        "Booking Status Distribution",
        "Success Rate by Vehicle Type",
        "Bookings by Hour of Day"
    )
)
```

```python
#plot 1:
fig.add_trace(
    go.Bar(
        x=vehicle_counts['Vehicle Type'],
        y=vehicle_counts['Booking ID'],
        name='Total Bookings',
        text=vehicle_counts['Booking ID'],
        textposition='auto'
    ),
    row=1, col=1
)


#plot 2:
fig.add_trace(
    go.Pie(
        labels=status_counts['Booking Status'],
        values=status_counts['Count'],
        hole=0.4
    ),
    row=1, col=2
)

#plot 3:
fig.add_trace(
    go.Bar(
        x=success_rate['Vehicle Type'],
        y=success_rate['Success Rate %'],
        name='Success Rate',
        text=success_rate['Success Rate %'].round(1),
        textposition='auto'
    ),
    row=2, col=1
)

#plot 4:
fig.add_trace(
    go.Scatter(
        x=hourly_bookings['Hour'],
        y=hourly_bookings['Booking ID'],
        mode='lines+markers',
        name='Hourly Bookings'
    ),
    row=2, col=2
)

#final step:
fig.update_layout(
    title_text="Uber NCR – Interactive Operations Dashboard",
    height=800,
    showlegend=False
)

fig.show()
```
